### Are SAT's biased?


One of the most controversial issues in the U.S. educational system is the efficacy of standardized tests, and whether they're unfair to certain groups. In this project, I will be investigating the correlations between SAT scores and demographics. I will correlate SAT scores with factors like race, gender, income, and more.


  The SAT, or Scholastic Aptitude Test, is an exam that U.S. high school students take before applying to college. Colleges take the test scores into account when deciding who to admit, so it's important to perform well on it.


  The test consists of three sections, each of which has 800 possible points. The combined score is out of 2,400 possible points. Organizations often rank high schools by their average SAT scores. The scores are also considered a measure of overall school district quality.

For this project I will use the following datasets published by New York city, these datasets cover demographic information and test scores. Here is the list of all the data sets used:

*	SAT scores by school - SAT scores for each high school in New York City
*	School attendance - Attendance information for each school in New York City
*	Class size - Information on class size for each school
*	AP test results - Advanced Placement (AP) exam results for each high school (passing an optional AP exam in a particular         subject can earn a student college credit in that subject)
*	Graduation outcomes - The percentage of students who graduated, and other outcome information
*	Demographics - Demographic information for each school
*	School survey - Surveys of parents, teachers, and students at each school

## Importing external data from NYC school survey taken by all parents, tecahers and students.

Every year, all parents, all teachers, and students in grades 6 - 12 take the NYC School Survey. The survey ranks among the largest surveys of any kind ever conducted nationally. Survey results provide insight into a school's learning environment and contribute a measure of diversification that goes beyond test scores on the Progress Report. NYC School Survey results contribute 10% - 15% of a school's Progress Report grade (the exact contribution to the Progress Report is dependant on school type). Survey questions assess the community's opinions on academic expectations, communication, engagement, and safety and respect. School leaders can use survey results to better understand their own school's strengths and target areas for improvement. The NYC School Survey helps school leaders understand what key members of the school community say about the learning environment at each school. The information captured by the survey is designed to support a dialogue among all members of the school community about how to make the school a better place to learn. New York City parents, students, and teachers reached new all-time participation highs on the 2011 School Survey. 960,191 surveys were submitted.

class_Size dataset does not have a DBN column as such, but comapring the calss_size and sat_results dataset it is observed that the DBN column is actually a combination of CSD and SCHOOL CODE column. The CSD column is padded though, so we can make necessary data manipulations to add a DBN column with appropriate value in the class_size dataset.

```python
# writing a custom function that can be supplied to the series.apply method to create a padded CSD column 
# in the class_size dataset.

def padding(num):
        num_str = str(num)
        length = len(num_str)
        if length < 2:
            num_str = num_str.zfill(2)
        elif length == 2:
            num_str = num_str
        return num_str
    

data["class_size"]["padded_csd"] = data["class_size"]["CSD"].apply(padding)
        
data["class_size"]["DBN"] = data["class_size"]["padded_csd"] + data["class_size"]["SCHOOL CODE"]

data["class_size"]["DBN"].head(5)
```

```python
data["hs_directory"]["DBN"] = data["hs_directory"]["dbn"]
```

In the below cell, we create an additional column in the sat_results dataset that adds up the scores for each section,

```python
# creating new columns
data["sat_results"]["SAT Math Avg. Score"] = pd.to_numeric(data["sat_results"]["SAT Math Avg. Score"] , errors = "coerce")

data["sat_results"]["SAT Critical Reading Avg. Score"] = pd.to_numeric(data["sat_results"]["SAT Critical Reading Avg. Score"] , errors = "coerce")

data["sat_results"]["SAT Writing Avg. Score"] = pd.to_numeric(data["sat_results"]["SAT Writing Avg. Score"] , errors = "coerce")

data["sat_results"]["sat_score"] = data["sat_results"]["SAT Math Avg. Score"] + data["sat_results"]["SAT Critical Reading Avg. Score"] + data["sat_results"]["SAT Writing Avg. Score"]


print(data["sat_results"]["sat_score"].head(5))
```

  Next, we'll want to parse the latitude and longitude coordinates for each school. This will enable us to map the schools and uncover any geographic patterns in the data. The coordinates are currently in the text field Location 1 in the hs_directory data set.

  We will parse extract the latitude and longitude values from the 'Location 1' column which has string values using regular expressions and add them as seperate columns in the hs_directory dataset.

```python
# creating functions to extract lattitude and longitude values using regular expressions
import re

# latitude
def extract_lat(string):
    coordinates = re.findall("\(.+\)", string)
    latitude = coordinates[0].split(",")[0].replace("(" , "")
    return latitude

data["hs_directory"]["lat"] = data["hs_directory"]["Location 1"].apply(extract_lat)

# longitude
def extract_lon(string):
    coordinates = re.findall("\(.+\)", string)
    longitude = coordinates[0].split(",")[1].replace(")" , "")
    return longitude

data["hs_directory"]["lon"] = data["hs_directory"]["Location 1"].apply(extract_lon)

# converting the lat and lon values to numeric
data["hs_directory"]["lat"] = pd.to_numeric(data["hs_directory"]["lat"] , errors = "coerce")
data["hs_directory"]["lon"] = pd.to_numeric(data["hs_directory"]["lon"] , errors = "coerce")



print(data["hs_directory"].head(5))
```

```python
class_size["CORE SUBJECT (MS CORE and 9-12 ONLY)"].unique()
```

```python
# aggregating the dataset at DBN level
grouped = class_size1.groupby("DBN")
class_size1 = grouped.agg(np.mean)
class_size1.reset_index(inplace = True)
data["class_size"] = class_size1
data["class_size"].head(5)
```

Now that we have manipulated all the required datasets for merging to have DBN as a unique value ( like PRIMARY KEY in SQL) we can combine(merge) all the datasets. 


```python
# combining the sat_results and the ap_test and graduation dataset. 
data["ap_2010"].info()
```

```python
data["graduation"].info()
```

```python
# using the left join to merge ap_2010 and graduation datasets to sat_resulats which will preserve all the DBN values from 
# sat_results dataset
combined = data["sat_results"]
combined = combined.merge(data["ap_2010"] , on = "DBN" , how = "left")
combined = combined.merge(data["graduation"] , on = "DBN" , how = "left")

combined.head(5)
```

```python
combined.shape
```

```python
# merging other datasets , Because these files contain information that's more valuable to our analysis and 
# also have fewer missing DBN values, we'll use the inner join type.

combined = combined.merge(data["class_size"] , on = "DBN" , how = "inner")
combined = combined.merge(data["demographics"] , on = "DBN" , how = "inner")
combined = combined.merge(data["survey"] , on = "DBN" , how = "inner")
combined = combined.merge(data["hs_directory"] , on = "DBN" , how = "inner")

combined.head(5)
```

```python
combined.shape
```

   After merging all the datasets, the number of rows are reduced from 479 to 363. This is because pandas couldn't find the DBN values that existed in sat_results in the other data sets. While this is worth investigating, we're currently looking for high-level correlations, so we don't need to dive into which DBNs are missing.

   Its noticed that we now have many columns with null (NaN) values. This is because we chose to do left joins, where some columns may not have had data. The data set also had some missing values to begin with. To fix this, we'll just fill in the missing values with the overall mean for the column. If there are Nan after that as well we will just fill those spaces with 0 , as we are going to perform correlation analysis later. 



```python
# filling the missing values with mean
mean = combined.mean()
combined = combined.fillna(mean)

combined = combined.fillna(0)
combined.head()
```

```python
# Extracting the school district from then DBN column to perform analysis on the district level
def extract_district(string):
    first_two = string[0:2]
    return first_two

combined["school_dist"] = combined["DBN"].apply(extract_district)
combined["school_dist"].head(5)
combined.head(5)
```

Now we have a cleaned dataset, we're now ready to analyze and visualize.
